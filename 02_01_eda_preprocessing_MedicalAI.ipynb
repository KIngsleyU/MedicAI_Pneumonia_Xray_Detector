```python
import glob
import random
import matplotlib.pyplot as plt
# import seaborne
from PIL import Image
import pandas as pd
```

```python
def get_random_image(dir, condition):
    placeholder = ""
    if condition == "n":
        placeholder = "NORMAL"
    elif condition == "p":
        placeholder = "PNEUMONIA"
    else:
        raise Exception("Invalid condition")
    folder = f"./data/chest_xray/{dir}/{placeholder}/*.jpeg"
    img_path = glob.glob(folder)
    max_lenghth = len(img_path)
    r_index = random.randint(0, max_lenghth)
    item = img_path[r_index]
    print(r_index, item)
    image = plt.imread(item)
    return plt.imshow(image)
```

```python
folder_train = "data/chest_xray/train/*/*.jpeg"

counter_p = 0
counter_n = 0

img_path_train = glob.glob(folder_train)
for i in img_path_train:
    img_train = Image.open(i)
    img_train = img_train.resize((224, 224))
    if "person" in i:
        plt.imsave(fname="./data/chest_xray/train/train_pneumonia"+str(counter_p)+".jpeg",
                   arr=img_train,
                   format="jpeg",
                   cmap="grey")
        counter_p += 1
    else:
        plt.imsave(fname="./data/chest_xray/train/train_normal"+str(counter_n)+".jpeg",
                   arr=img_train,
                   format="jpeg",
                   cmap="grey")
        counter_n += 1
        
```
